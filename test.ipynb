```python
# --- Step 2: Define the Preprocessing Function ---
# This function will take a batch of songs and chop their long duration lists
# into fixed-length chunks that the model can handle.

SEQUENCE_LENGTH = 128 

def chunk_examples(examples):
    # Dictionaries to hold the new, chunked data
    chunked_inputs = {
        'original_durations': [],
        'quantized_durations': [],
        'singer_id': [],
    }
    
    # Iterate through each song in the batch
    for i in range(len(examples["durations"])):
        durs = examples["durations"][i]
        q_durs = examples["quantized_durations"][i]
        sid = examples["singer_id"][i]
        
        # Chop the long lists into smaller, fixed-length chunks
        for j in range(0, len(durs) - SEQUENCE_LENGTH, SEQUENCE_LENGTH):
            chunked_inputs['original_durations'].append(durs[j : j + SEQUENCE_LENGTH])
            chunked_inputs['quantized_durations'].append(q_durs[j : j + SEQUENCE_LENGTH])
            chunked_inputs['singer_id'].append(sid)
            
    return chunked_inputs


# --- Step 3: Apply the Preprocessing ---
# We use .map() to apply our function to the entire dataset.
# `batched=True` sends multiple rows at a time to our function for efficiency.
# `remove_columns` gets rid of the old, variable-length columns.
print("Chunking dataset into fixed-length sequences...")
processed_dataset = train_dataset.map(
    chunk_examples,
    batched=True,
    remove_columns=train_dataset.column_names
)
```

```python
# --- Step 4: Set Format and Create DataLoaders ---
# Now we format the dataset to output PyTorch tensors and create the DataLoaders
# which will handle batching and shuffling for us during training.

BATCH_SIZE = 32

# --- Dataloader for Model A (Control) ---
# Goal: original -> original
# We need to duplicate the 'original_durations' column to use it for both input and labels.
model_A_dataset = processed_dataset.map(
    lambda batch: {"labels": batch["original_durations"]},
    batched=True
)
# Now we have: 'original_durations', 'quantized_durations', 'singer_id', 'labels'

# Rename 'original_durations' to 'input_ids' for the model's input
model_A_dataset = model_A_dataset.rename_column("original_durations", "input_ids")
model_A_dataset.set_format(type='torch', columns=['input_ids', 'labels', 'singer_id'])
model_A_dataloader = DataLoader(model_A_dataset, batch_size=BATCH_SIZE, shuffle=True)
print("DataLoader for Control Model (original -> original) created.")


# --- Dataloader for Model B (Your Method) ---
# Goal: quantized -> original
model_B_dataset = processed_dataset.map(
    lambda batch: {"labels": batch["original_durations"]},
    batched=True
)

# Rename 'quantized_durations' to 'input_ids' for the model's input
model_B_dataset = model_B_dataset.rename_column("quantized_durations", "input_ids")
model_B_dataset.set_format(type='torch', columns=['input_ids', 'labels', 'singer_id'])
model_B_dataloader = DataLoader(model_B_dataset, batch_size=BATCH_SIZE, shuffle=True)
print("DataLoader for Your Method's Model (quantized -> original) created.")
```

```python
# --- Step 5: Verify One Batch ---
# Let's pull one batch from the dataloader for Model B to see what it looks like.
# This is the data in its final form, "right before being input into model".

print("\n--- Verifying one batch for Model B (Your Method) ---")
one_batch = next(iter(model_B_dataloader))

# The dataloader gives us a dictionary of tensors
input_tensor = one_batch['input_ids']
label_tensor = one_batch['labels']
sid_tensor = one_batch['singer_id']

print(f"Shape of the input tensor: {input_tensor.shape}")
print(f"Shape of the label tensor: {label_tensor.shape}")
print(f"Shape of the singer_id tensor: {sid_tensor.shape}")
print("\nFirst input sequence in the batch (quantized):")
print(input_tensor[0])
print("\nCorresponding label sequence in the batch (original):")
print(label_tensor[0])
print("\nCorresponding singer_id:")
print(sid_tensor[0])
```
